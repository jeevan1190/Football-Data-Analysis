# Scatter Plot:

Medals vs Event

# Pairplot:

```python
team_goals = pd.concat([
    df.groupby('home_team')['fulltime_home'].sum(),
    df.groupby('away_team')['fulltime_away'].sum()
]).groupby(level=0).sum().sort_values(ascending=False).head(5)

# Prepare dataframe for pairplot
df_goals = pd.DataFrame({
    'Team': team_goals.index,
    'Goals_Home': [df[df['home_team']==t]['fulltime_home'].sum() for t in team_goals.index],
    'Goals_Away': [df[df['away_team']==t]['fulltime_away'].sum() for t in team_goals.index],
})

# Total goals
df_goals['Total_Goals'] = df_goals['Goals_Home'] + df_goals['Goals_Away']

sns.pairplot(df_goals, vars=['Goals_Home','Goals_Away','Total_Goals'], hue='Team')
plt.show()
```

# Box Plot:

Top 5 Countries with No Gold Medals

```python
top_teams = pd.concat([df['home_team'], df['away_team']]).value_counts().head(5).index

# Prepare data for boxplot: total goals in matches for top teams
df_box = pd.DataFrame({'Team': [], 'Goals': []})

for t in top_teams:
    home_goals = df[df['home_team']==t]['fulltime_home']
    away_goals = df[df['away_team']==t]['fulltime_away']
    
    # Concatenate Series using pd.concat
    all_goals = pd.concat([home_goals, away_goals])
    
    temp = pd.DataFrame({
        'Team': [t]*len(all_goals),
        'Goals': all_goals.values
    })
    df_box = pd.concat([df_box, temp], ignore_index=True)

# Boxplot
plt.figure(figsize=(14,7))
sns.boxplot(data=df_box, x='Team', y='Goals')
plt.title("Goals Distribution of Top 5 Teams")
plt.show()
```

# Heatmap:

Count Of Matches Hosted by Each Home Team

```python
host_counts = df['home_team'].value_counts().reset_index()
host_counts.columns = ['Team','Matches_Hosted']

plt.figure(figsize=(12,4))
sns.heatmap(host_counts.set_index('Team').T, annot=True, cmap='coolwarm', cbar=True)
plt.title("Matches Hosted by Teams (as Home Team)")
plt.show()
```

# Pie Chart:

Top Six Teams For Pie Chart

```python
# Total goals per team (home + away)
team_goals = pd.concat([
    df.groupby('home_team')['fulltime_home'].sum(),
    df.groupby('away_team')['fulltime_away'].sum()
]).groupby(level=0).sum()

# Prepare top 6 teams for pie chart
top_teams = team_goals.sort_values(ascending=False).head(6)

plt.figure(figsize=(8,8))
plt.pie(top_teams, labels=top_teams.index, autopct='%1.1f%%', startangle=140)
plt.title("Team-wise Total Goals (Top 6 Teams)")
plt.show()
```

# Violin Plot:

Team Wise Goals Distributions

```python
# Top 5 teams by total goals
top_teams = pd.concat([
    df.groupby('home_team')['fulltime_home'].sum(),
    df.groupby('away_team')['fulltime_away'].sum()
]).groupby(level=0).sum().sort_values(ascending=False).head(5).index

# Prepare data for violinplot
data_list = []
for t in top_teams:
    home_goals = df[df['home_team']==t]['fulltime_home']
    away_goals = df[df['away_team']==t]['fulltime_away']
    all_goals = pd.concat([home_goals, away_goals])
    temp = pd.DataFrame({'Team':[t]*len(all_goals), 'Goals':all_goals.values})
    data_list.append(temp)

df_violin = pd.concat(data_list, ignore_index=True)

plt.figure(figsize=(12,6))
sns.violinplot(data=df_violin, x='Team', y='Goals')
plt.title("Goals Distribution by Top 5 Teams")
plt.show()
```

# KDE Plot:

Season Wise Total Goal Distribution

```python
# Extract season start year (numeric)
df['season_start'] = df['season'].str[:4].astype(int)

# Define 2 groups: odd vs even season year as example (analogous to Summer/Winter)
df['Season_Type'] = df['season_start'].apply(lambda x: 'Odd Season' if x%2 else 'Even Season')

# Calculate total goals per match
df['Total_Goals'] = df['fulltime_home'] + df['fulltime_away']

plt.figure(figsize=(12,6))
sns.kdeplot(data=df, x='Total_Goals', hue='Season_Type', fill=True)
plt.title("Total Goals Distribution: Odd vs Even Seasons")
plt.show()
```

# Histogram:

 Total Goals Per Match

```python
# Total goals per match
df['Total_Goals'] = df['fulltime_home'] + df['fulltime_away']

plt.figure(figsize=(10,6))
sns.histplot(df['Total_Goals'], bins=10, kde=True, color='skyblue')
plt.title("Total Goals Distribution per Match")
plt.xlabel("Number of Goals")
plt.ylabel("Match Count")
plt.show()
```

# Donut Chart:

Home vs Away Goals Distributions

```python
# Total goals as home vs away
home_goals = df['fulltime_home'].sum()
away_goals = df['fulltime_away'].sum()

goals = [home_goals, away_goals]
categories = ['Home Goals', 'Away Goals']

plt.figure(figsize=(8,8))
plt.pie(goals, labels=categories, autopct="%1.1f%%", startangle=90)
centre = plt.Circle((0,0),0.70, fc="white")
fig = plt.gcf()
fig.gca().add_artist(centre)
plt.title("Goal Distribution: Home vs Away")
plt.show()
```

# TreeMap(Squarify):

Top 20 Teams By Total Goals 

```python
# Total goals per team (home + away)
team_goals = pd.concat([
    df.groupby('home_team')['fulltime_home'].sum(),
    df.groupby('away_team')['fulltime_away'].sum()
]).groupby(level=0).sum().sort_values(ascending=False).head(20).reset_index()

team_goals.columns = ['Team','Total_Goals']

plt.figure(figsize=(16,9))
squarify.plot(
    sizes=team_goals["Total_Goals"],
    label=team_goals["Team"]+" ("+team_goals["Total_Goals"].astype(str)+")",
    color=plt.cm.Set3.colors,
    alpha=0.9,
    pad=True,
    text_kwargs={'fontsize':12, 'weight':'bold'}
)
plt.axis("off")
plt.title("Top 20 Teams by Total Goals", fontsize=18, weight='bold')
plt.show()
```
